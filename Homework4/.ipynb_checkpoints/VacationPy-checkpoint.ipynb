```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os

# Import API key
from api_keys import gkey
```

```python
file='weather_data.csv'
part_one_df = pd.read_csv(file)
part_one_df.head(100)
```

```python
gmaps.configure(api_key=gkey)
```

```python
locations = part_one_df[["Lat", "Lng"]].astype(float)
Humidity=part_one_df["Humidity"].astype(int)
Humidity.max()
```

```python
fig = gmaps.figure()

heat_layer = gmaps.heatmap_layer(locations, weights=Humidity, 
                                 dissipating=False, max_intensity=250,
                                 point_radius = 4)

fig.add_layer(heat_layer)

fig
```

```python
part_one_df_select = part_one_df.loc[(part_one_df["Max Temp"] <= 300 ) , :]
part_one_df_select = part_one_df_select.loc[(part_one_df["Wind Speed"] == 1 ) , :]
hotel = part_one_df_select.loc[(part_one_df["Cloudiness"] <= 20 ) , :]
hotel
```

```python
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
```

```python
#str(list1).replace('[','"').replace(']','"') 
target_coordinates = hotel[["Lat","Lng"]].values.tolist()
```

```python
hname=[]
hlat=[]
hlng=[]
vicinity=[]
for r in target_coordinates:
    t_coordinates=str(r).replace('[','').replace(']','') 
    t_radius = 5000
    t_type = "hotel"
    params = {
    "location": t_coordinates,
    "radius": t_radius,
    "type": t_type,
    "key": gkey}
    try:
        response = requests.get(base_url, params=params).json()
        hname.append(response['results'][1]['name'])
        hlat.append(response['results'][1]['geometry']['location']['lat'])
        hlng.append(response['results'][1]['geometry']['location']['lng'])
        vicinity.append(response['results'][1]['vicinity'])
    except:
        print("Hotel not found. Skipping...")
    continue


```

```python
dic={"Name": hname,
    "Lat":hlat, 
    "Lng":hlng, 
    "City": vicinity}
hotel_df = pd.DataFrame(dic)
hotel_df.head()
```

```python
coordinates =hotel_df[["Lat","Lng"]]
namesH=hotel_df["Name"]
```

```python
figure_layout = {
    'width': '400px',
    'height': '300px',
    'border': '1px solid black',
    'padding': '1px',
    'margin': '0 auto 0 auto'
}
fig = gmaps.figure(layout=figure_layout)
markers = gmaps.marker_layer(coordinates,  info_box_content=namesH)
fig.add_layer(markers)
fig.add_layer(heat_layer)
fig
```
